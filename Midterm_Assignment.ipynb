#Poverty and Disability: US vs. The World

The US

```python
import pandas as pd

file_path = '/content/Disability Prevalence by Disability Type in the United States  Northeast ADA Center.csv'
df = pd.read_csv(file_path)
display(df.head())
```

```python
# First, clean the 'Estimate (#)' column to remove commas and convert it to numeric
df['Estimate (#)'] = df['Estimate (#)'].str.replace(',', '', regex=False).astype(int)

# Calculate the average 'Estimate (#)' for each 'Disability Type'
average_estimate_by_disability_type = df.groupby('Disability Type')['Estimate (#)'].mean().reset_index()

# Display the result
display(average_estimate_by_disability_type)
```

```python
# The 'Estimate (#)' column is already numeric, so no cleaning is needed.
# Calculate the average 'Estimate (#)' for each 'Disability Type'
average_estimate_by_disability_type = df.groupby('Disability Type')['Estimate (#)'].mean().reset_index()

# Display the result
display(average_estimate_by_disability_type)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a bar chart for 'Disability Type' vs 'Estimate (#)' (base population)
plt.figure(figsize=(12, 8))
sns.barplot(x='Estimate (#)', y='Disability Type', data=average_estimate_by_disability_type.sort_values(by='Estimate (#)', ascending=False), palette='viridis', hue='Disability Type', legend=False)
plt.title('Average Base Population by Disability Type', fontsize=16)
plt.xlabel('Average Estimate (Number of Individuals)', fontsize=12)
plt.ylabel('Disability Type', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Clean 'Estimate (%)' column by removing '%' and converting to numeric
df['Estimate (%)'] = df['Estimate (%)'].str.replace('%', '', regex=False).astype(float)

# Sort data by prevalence for better visualization
df_sorted = df.sort_values(by='Estimate (%)', ascending=False)
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x='Estimate (%)', y='Disability Type', data=df_sorted, palette='viridis', hue='Disability Type', legend=False)
plt.title('Prevalence of Different Disability Types in the United States')
plt.xlabel('Prevalence (%)')
plt.ylabel('Disability Type')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create the pie chart
plt.figure(figsize=(10, 8))
plt.pie(average_estimate_by_disability_type['Estimate (#)'],
        labels=average_estimate_by_disability_type['Disability Type'],
        autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Average Estimate (#) by Disability Type')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
url = 'https://www.disabilitydatainitiative.org/country-briefs/kh/#2-table-1-cambodia-prevalence-of-functional-difficulties'
response = requests.get(url, verify=False)
print(f"Response Status Code: {response.status_code}")
```

```python
import pandas as pd

tables = pd.read_html(response.text)

for i, df_table in enumerate(tables):
    print(f"Table {i+1} Head:")
    display(df_table.head())
    print(f"Table {i+1} Info:")
    df_table.info()
    print("\n" + "-"*30 + "\n")
```

```python
df_prevalence = tables[0]
df_socio_economic = tables[1]

print("Summary for Disability Prevalence by Group (df_prevalence):")
display(df_prevalence.head())
print(df_prevalence.info())
print("\n" + "-"*50 + "\n")

print("Summary for Socio-Economic Indicators by Difficulty Level (df_socio_economic):")
display(df_socio_economic.head())
print(df_socio_economic.info())
```

```python
df_prevalence_melted = pd.melt(df_prevalence, id_vars=['Group'], value_vars=['Any difficulty', 'Some difficulty', 'At least a lot of difficulty'], var_name='Difficulty Level', value_name='Prevalence')
display(df_prevalence_melted.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 8))
sns.barplot(x='Group', y='Prevalence', hue='Difficulty Level', data=df_prevalence_melted, palette='viridis')
plt.title('Disability Prevalence by Group and Difficulty Level', fontsize=16)
plt.xlabel('Demographic Group', fontsize=12)
plt.ylabel('Prevalence (%)', fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotate and align x-axis labels
plt.legend(title='Difficulty Level')
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
print("First few rows of df_socio_economic:")
display(df_socio_economic.head())

print("\nSummary of df_socio_economic:")
df_socio_economic.info()
```

```python
df_socio_economic['Difference between no difficulty and some difficulty'] = df_socio_economic['Difference between no difficulty and some difficulty'].astype(str).str.replace(r'[^\d.-]+', '', regex=True)
df_socio_economic['Difference between no difficulty and some difficulty'] = pd.to_numeric(df_socio_economic['Difference between no difficulty and some difficulty'], errors='coerce').astype(float)

df_socio_economic['Difference between no difficulty and at least a lot of difficulty'] = df_socio_economic['Difference between no difficulty and at least a lot of difficulty'].astype(str).str.replace(r'[^\d.-]+', '', regex=True)
df_socio_economic['Difference between no difficulty and at least a lot of difficulty'] = pd.to_numeric(df_socio_economic['Difference between no difficulty and at least a lot of difficulty'], errors='coerce').astype(float)

print("Info after cleaning and conversion:")
display(df_socio_economic.info())
```

```python
df_socio_economic_melted = pd.melt(df_socio_economic,
                                      id_vars=['Indicator'],
                                      value_vars=[
                                          'No Difficulty',
                                          'Some Difficulty',
                                          'Difference between no difficulty and some difficulty',
                                          'At least a lot of difficulty',
                                          'Difference between no difficulty and at least a lot of difficulty'
                                      ],
                                      var_name='Difficulty Level',
                                      value_name='Value')

print("Melted DataFrame Head:")
display(df_socio_economic_melted.head())
print("Melted DataFrame Info:")
display(df_socio_economic_melted.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 10))
sns.barplot(x='Indicator', y='Value', hue='Difficulty Level', data=df_socio_economic_melted, palette='viridis')
plt.title('Socio-Economic Indicators by Difficulty Level', fontsize=16)
plt.xlabel('Socio-Economic Indicator', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotate and align x-axis labels
plt.legend(title='Difficulty Level', bbox_to_anchor=(1.05, 1), loc='upper left') # Place legend outside the plot
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
